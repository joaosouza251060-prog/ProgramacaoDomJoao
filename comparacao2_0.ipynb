```python
print("Comparando os números")

N1 = float(input("Digite o primeiro número: "))
N2 = float(input("Digite o segundo número: "))

if N1 > N2:
  print(f"{N1} é maior que {N2}")
elif N1 < N2:
  print(f"{N1} é menor que {N2}")
else:
  print(f"{N1} é igual a {N2}")
```
